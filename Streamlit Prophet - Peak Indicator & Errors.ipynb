```python
# Update Today Date
Today=datetime.datetime.now()
# Yesterday=datetime.datetime.now()- datetime.timedelta(days=1)
# print(Yesterday)
# print(Today)
```

```python
#Download from Yahoo finance
ticker="CT=F"
period1=int(time.mktime(datetime.datetime(2000,3,1,23,59).timetuple()))
period2=int(time.mktime(Today.timetuple())) # To update with latest date
interval='1d' #1wk,1m

query_string=f'https://query1.finance.yahoo.com/v7/finance/download/{ticker}?period1={period1}&period2={period2}&interval={interval}&events=history&includeAdjustedClose=true'

dfy=pd.read_csv(query_string,parse_dates=['Date'],index_col="Date")
```

```python
dfy.head()
```

```python
#drop null values
dfy=dfy.dropna()
dfy=dfy.drop_duplicates(subset=None, keep='first', inplace=False, ignore_index=False)
```

```python
dfy.describe()
```

```python
# updating our dataFrame to have only
# one column 'Adj Close' as rest all columns
# are of no use for us at the moment
# using .to_frame() to convert pandas series
# into dataframe.
dfya = dfy['Adj Close'].to_frame()
  
# calculating simple moving average
# using .rolling(window).mean() ,
# with window size = 30
dfya['SMA30'] = dfya['Adj Close'].rolling(30).mean()

# removing all the NULL values using 
# dropna() method
dfya.dropna(inplace=True)
  
# printing Dataframe
dfya
```

```python
# Define PEAK COL
def P_Index(row):
    if row['Adj Close'] > row['SMA30']+2:
        return 1
    elif row['Adj Close'] < row['SMA30']-5 :
        return -1
    else:
        return 0
```

```python
dfya['P_INDEX']=dfya.apply(P_Index, axis=1)
```

```python
dfya.tail(20)
```

```python
  
# calculating simple moving average
# using .rolling(window).mean() ,
# with window size = 30
dfya['SMA_P7'] = dfya['P_INDEX'].rolling(7).mean()

dfya.dropna(inplace=True)
  
# printing Dataframe
# dfya
```
